# **Matplotlib**

````python
import matplotlib.pyplot as plt

# x축 입력값 [1,2,3], y축 입력값 [2,4,6]으로 선 그래프 생성
plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
````

## **figure와 axis의 이해 및 활용**

* figure : 그림을 그리기 위한 캔버스, 실제 그림을 그리는 역할은 아님
* axes : 실제 그림을 그리는 메서드들을 가짐!

![figure and axes](https://wikidocs.net/images/page/4763/r15.10.png)

````python
plt.figure(figsize=(10,4))

plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
````

````python
figure = plt.figure(figsize=(10,4))
print(type(figure))
````

````python
#figure가 그림 그리는 역할은 아니지만 그림판의 배경 색깔을 바꾸는 등의 설정은 가능
#axes가 차지하는 영역까지는 칠하지 않음, figure 위에 axes가 그려지는 느낌
plt.figure(figsize=(4,4), facecolor='yellow')
plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
````

````python
ax = plt.axes()
print(type(ax))
````

pyplot 모듈이 import 되어있는 상태에서는 plt.show()를 입력하지 않더라도 axes가 할당되어 있으면 자동으로 그림을 표출됨

````python
# pyplot에서 설정된 figure와 axes 객체를 함께 가져오기
fig, ax = plt.subplots()
print('fig type:',fig, '\nax type:',type(ax))
````

````python
fig, ax = plt.subplots(figsize=(4,4))

#axes 객체의 plot을 이용하여 선 그래프 그리기
ax.plot([1,2,3],[2,4,6])

#axes 객체의 set_title()을 이용하여 title 지정
ax.set_title('Hello plot')
plt.show()
````

### **xlabel, ylabel**

### **xticks, yticks**

* 축의 눈금값(틱값)을 회전시키는 함수
* 축의 값이 문자열이고 많은 눈금값이 있을 경우 틱의 회전이 필요하기도 함

````python
x_value = np.arange(0,100)    # 0~99
y_value = 2*x_value

plt.plot(x_value,y_value)
plt.xlabel('x axis'); plt.ylabel('y axis')

#x축 눈금을 90도 회전
plt.xticks(rotation=90)
plt.show()
````

* 축 값이 문자열이 아닌 숫자열인 경우에는 자동적으로 최솟값과 최댓값, 축의 크기에 따라 눈금 간격을 만들어줌
* 눈금 간격을 지정하고 싶다면 xticks() 함수의 인자로 ticks 값을 바꿔주면 됨

````python
x_value = np.arange(0,100)
y_value = 2*x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis'); plt.ylabel('y axis')
plt.title('Hello plot')

plt.xticks(ticks=np.arange(0,100,5),rotation=90)
plt.show()
````

### **legend**

* plt.plot() 함수 내의 label 인자값을 설정하여 해당 plot에 명칭을 부여함
* plt.legend() 를 호출하여 axes 상에 나타냄

````python
x_value_01 = np.arange(0,100)
y_value_01 = 2*x_value_01
y_value_02= 4*x_value_01

#plt.plot(lable=' ')을 통해 각 plot에 명칭을 부여
plt.plot(x_value_01, y_value_01, color='green', label='green line')
plt.plot(x_value_01, y_value_02, color='red', label='red line')

#plt.legend() 를 호출해야 그림에서 legend 확인 가능
plt.legend()

plt.xlabel('x axis'); plt.ylabel('y axis')
````

````python
#선과 막대그래프 같이 그리기
x_value_01 = np.arange(1,10)
y_value_01 = 2*x_value_01

plt.plot(x_value_01, y_value_01,color='red',marker='o',linestyle='dashed',label='red line')
plt.bar(x_value_01, y_value_01, color='green',label='bar plot')
plt.legend()

plt.xlabel('x axis'); plt.ylabel('y axis')
plt.title('Hello plot')

plt.show()
````

axes 매서드를 활용해 같은 결과 도출하기

````python
figure = plt.figure()

#axes 객체를 추출하여 ax 변수에 할당
ax = plt.axes()

ax.plot(x_value_01, y_value_01, color='red', linestyle='dashed', marker='o', label='red line')
ax.bar(x_value_01, y_value_01, color='green', label='bar plot')
ax.legend()

ax.set_xlabel('x axis'); ax.set_ylabel('y axis')
ax.set_title('Hello plot')

plt.show()
````

## **여러 개의 subplots들을 이용해 개별 subplot별로 시각화하기**

* 개별 Axes 객체를 직접 이용해서 시각화를 구현해야함!!

````python
#axes를 배열 형태로 받음
#ax[0] : 왼쪽에 생성된 Axes 객체, ax[1] : 오른쪽에 생성된 Axes 객체를 가리킴
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,3))
````

````python
print('ax type:', type(ax))          #numpy의 ndarray
print('ax[0] type:', type(ax[0]))    #Axes
print('ax[1] type:', type(ax[1]))
````

````python
x_value_01 = np.arange(1,10)
x_value_02 = np.arange(1,20)
y_value_01 = 2*x_value_01
y_value_02 = 2*x_value_02

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,3))

#개별 axes 객체들에 그래프를 그림
ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].bar(x_value_02, y_value_02, color='green',label='bar plot')
#개별 axes 객체 각각에 범례를 설정
ax[0].legend(); ax[1].legend()

ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')

ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')

plt.show()
````

**nrows>=2, ncols>=2인 경우**

* 2차원 배열 인덱싱 방식으로 Axes 객체에 접근해야함
    * ax[ i ][ j ]

````python
#nrow>=2, ncol>=2인 경우

x_value_01 = np.arange(1,10)
x_value_02 = np.arange(1,20)
y_value_01 = 2*x_value_01
y_value_02 = 2*x_value_02

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green',label='green bar')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', label='green line')
ax[1][1].bar(x_value_02, y_value_02, color='red',label='red bar')
ax[0][0].legend();ax[0][1].legend();ax[1][0].legend();ax[1][1].legend()

ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

plt.show()
````
