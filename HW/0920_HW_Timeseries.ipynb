## **datetime 오브젝트**

* datetime 라이브러리 : 날짜와 시간을 처리하는 등의 기능을 제공
* date, time, datetime(날짜+시간) 오브젝트 등이 포함되어있음

````python
#datetime library import
#import datetime만 한다면 불러올 때 datetime.datetime 으로 코드를 써야함
from datetime import datetime
````

````python
#now() 메서드를 이용하여 현재시각 출력 (단, UTC를 기준으로 함, 한국은 UTC+9)
now1 = datetime.now()
print(now1)
````

````python
#today() 메서드로도 같은 결과
now2 = datetime.today()
print(now2)
````

````python
#datetime.now()에서 원하는 시간 정보만 추출하기
#datetime.now().strftime()

now = datetime.now()
print(now)

nowDate = now.strftime('%Y-%m-%d')
print(nowDate)

nowTime = now.strftime('%H:%M:%S')
print(nowTime)

nowDatetime = now.strftime('%Y-%m-%d %H:%M:%S')
print(nowDatetime)
````

````python
#시간을 직접 입력하여 datetime으로 오브젝트를 생성
t1 = datetime.now()
t2 = datetime(1970,1,1)   #year, month, day
t3 = datetime(1970, 12, 12, 13, 24, 34)  #year, month, day, hour, minute, second

print(t1); print(t2); print(t3)
````

````python
#두 datetime 오브젝트의 차이
diff1 = t1-t2
print(diff1)
print(type(diff1))
````

````python
diff2 = t2-t1
print(diff2)
print(type(diff2))
````

````python
ebola = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ESAA/data/country_timeseries.csv', parse_dates=['Date'])
ebola.iloc[:5,:5]
````

````python
ebola.tail().iloc[:,:5]
````

````python
ebola.index = ebola['Date']
new_idx = pd.date_range(ebola.index.min(), ebola.index.max())
#ebola 데이터(역순)와 달리 시간 순서대로 생성되어있음
new_idx
````

````python
new_idx = reversed(new_idx)
new_idx
````

````python
#2015-01-01 데이터가 생김
ebola = ebola.reindex(new_idx)
ebola.head().iloc[:,:5]
````

````python
ebola.tail().iloc[:,:5]
````

````python
# 각 나라의 에볼라 발병일 옮기기
# last_valid_index : 가장 오래된 데이터를 가져옴
# first_valid_index : 가장 최근의 데이터를 가져옴
last_valid = ebola.apply(pd.Series.last_valid_index)
print(last_valid)
````

````python
first_valid = ebola.apply(pd.Series.first_valid_index)
print(first_valid)
````

````python
earliest_date = ebola.index.min()
print(earliest_date)
````

````python
shift_values = last_valid - earliest_date
print(shift_values)
````

````python
# shift_values 값만큼 옮김
ebola_dict = {}
for idx, col in enumerate(ebola):
    d = shift_values[idx].days
    shifted = ebola[col].shift(d)
    ebola_dict[col] = shifted
````

````python
# 에볼라의 최초 발병일(14.03.22)을 기준으로 모든 열의 데이터가 옮겨짐
ebola_shift = pd.DataFrame(ebola_dict)
ebola_shift.tail()
````

````python
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Date','Day'], axis=1)
ebola_shift.tail()
````

````python
fig, ax = plt.subplots()
ax = ebola_shift.plot(ax=ax)
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show()
````
